```python
import pandas as pd
```

```python
df_orig_Alan  = pd.read_json('myfileAlan.json') # load data in
df_Alan= pd.DataFrame(list(df_orig_Alan.sensors)) # get data
df_Alan['timestamp'] = df_orig_Alan.timestamp
df_Alan = df_Alan.set_index('timestamp')
df_Alan
```

```python
df_orig_Odin  = pd.read_json('myfileOdin.json') # load data in
df_Odin= pd.DataFrame(list(df_orig_Odin.sensors)) # get data
df_Odin['timestamp'] = df_orig_Odin.timestamp
df_Odin = df_Odin.set_index('timestamp')
df_Odin
```

```python
fig = df_Alan.plot(kind='line', title=['Conductivity','Dissolved Oxygen','Salinity','Temperature','Turbidly','pH'],
        figsize=(20, 16), fontsize=13, subplots=True,grid=True)
```

```python
fig = df_Odin.plot(kind='line',
        figsize=(20, 16), fontsize=13, subplots=True,grid=True)
```

```python
combined = pd.concat([df_Alan, df_Odin], axis=1)
combined = combined.dropna()
combined = combined.drop([
       'DistLightning', 'LightningStrikes',
       'TiltNS', 'TiltWE', 'Rain', 'MaxWindSpeed',
        'WindDir', 'WindSpeed','AirTemp', 'BaroPressure','RelHumidTemp', 'SolarFlux', 'SolarTotalFlux',
       'VaporPressure'], axis=1)
fig = combined.plot(kind='line',
        figsize=(20, 50), fontsize=20, subplots=True,grid=True)
combined.columns
```
